```python
# Exercise 1: Find All Pairs of Elements That Sum to a Target
arr = [1, 2, 3, 4, 2, 5, 7, 2]
target_sum = 7

# Initialize a list to store the pairs
result = []

# Loop through the array to find all pairs
for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] + arr[j] == target_sum:
            result.append([arr[i], arr[j]])

# Print the result
print(result)


# Exercise 2: String Transformation
def accum(s):
    result = []
    for i in range(len(s)):
        result.append(s[i].upper() + s[i].lower() * i)
    return '-'.join(result)

# Exercise 3: Remove First and Last Characters
def remove_char(s):
    return s[1:-1]

# Exercise 4: Repeat String
def repeat_str(n, s):
    result=n*s
    return result

# Exercise 5: Multiply Array Elements
def multiply_elements(arr):
    result = 1
    for i in range(len(arr)):
        result =result*arr[i]
        if i==len(arr)-1:
            print(str(arr[i])+"="+str(result),end="")
        else:
            print(str(arr[i])+"*",end="")

# Exercise 6: Remove Duplicates
def remove_duplicates(lst):
    unique_elements = []
    for item in lst:
        if item not in unique_elements:
            unique_elements.append(item)
    return unique_elements

# Exercise 7: Find Positions of an Element
def find_positions(lst, target):
    positions = []
    for index, value in enumerate(lst):
        if value == target:
            positions.append(index)
    return positions

# Exercise 8: Lottery Hits
def lottery_hits(drawn, bets):
    hits = 0
    for number in bets:
        if number in drawn:
            hits += 1
    return hits

# Exercise 9: Check Palindrome Number
def is_palindrome(n):
    original = n
    reverse = 0
    while n > 0:
        digit = n % 10
        reverse = reverse * 10 + digit
        n = n // 10
    return original == reverse

# Exercise 10: Print Pattern
def print_pattern(rows):
    for i in range(1, rows + 1):
        print('* ' * i)
    for i in range(rows - 1, 0, -1):
        print('* ' * i)

# Exercise 11: Concatenate two lists index-wise
def concatenate_lists(list1, list2):
    return [x + y for x, y in zip(list1, list2)]

# Example usage for Exercise 1:
original_matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(transpose(original_matrix))

# Example usage for Exercise 2:
print(accum("abcd"))      # "A-Bb-Ccc-Dddd"

# Example usage for Exercise 3:
print(remove_char('eloquent'))  # 'loquen'

# Example usage for Exercise 4:
print(repeat_str(4, 'a'))       # 'aaaa'

# Example usage for Exercise 5:
print(multiply_elements([1, 2, 3, 4]))  # 24

# Example usage for Exercise 6:
my_list = [10, 8, 8, 8, 6, 6, 4, 2, 100, 100, 100]
print(remove_duplicates(my_list))  # [10, 8, 6, 4, 2, 100]

# Example usage for Exercise 7:
my_list = [0, 3, 12, 12, 8, 2]
input_value = 12
print(find_positions(my_list, input_value))  # [2, 3]

# Example usage for Exercise 8:
drawn = [5, 11, 9, 42, 3, 49]
bets = [3, 7, 11, 42, 34, 49]
print(lottery_hits(drawn, bets))  # 4

# Example usage for Exercise 9:
print(is_palindrome(121))  # True

# Example usage for Exercise 10:
print_pattern(5)

# Example usage for Exercise 11:
list1 = ["M", "na", "i", "Ke"]
list2 = ["y", "me", "s", "lly"]
print(concatenate_lists(list1, list2))  # ['My', 'name', 'is', 'Kelly']
```
